```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('mobile_data_flipkart_scraping.csv')
```

```python
df.index=list(range(1,468))
```

```python
df
```

```python
budget_10k=[]
for i in df['Price']:
    if i<=10000:
        budget_10k.append('Yes')
    else:
        budget_10k.append('No')
```

```python
budget_50k=[]
for i in df['Price']:
    if i>=50000:
        budget_50k.append('Yes')
    else:
        budget_50k.append('No')
```

```python
df['Budget_below_10k']=budget_10k
df['Budget_above_50k']=budget_50k
```

```python
df.plot.scatter('Price','Current_Offer')
```
