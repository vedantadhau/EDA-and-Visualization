```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.style
```

```python
df=pd.read_csv('final_without_index.csv',index_col=0)
```

```python
df.isnull().sum()
```

```python
df.dropna(how='any',subset=['Size_in_inches','Warranty_for_handset_in_years'],inplace=True)
```

```python
df
```

## UNIVARIATE  PLOT

```python
plt.figure(figsize=(20,13))
plt.style.use('ggplot')

plt.subplot(331)
plt.plot(df.Price)
plt.title('Prices')

plt.subplot(332)
plt.plot(df.Ratings)
plt.ylim(3,5)
plt.title('Ratings')

plt.subplot(333)
plt.plot(df.Current_Offer)
plt.title('Offers')

plt.subplot(334)
plt.plot(df['RAM in GB'])
plt.title('RAM')

plt.subplot(335)
plt.plot(df['Storage in GB'])
plt.title('ROM')

plt.subplot(336)
plt.plot(df['Warranty_for_accessories_in_month'])
plt.title('Warranty_for_accessories')

plt.subplot(337)
plt.plot(df['Battery_in_mAh'])
plt.title('Battery')

plt.subplot(338)
plt.plot(df['Warranty_for_handset_in_years'])
plt.title('Warranty_for_handset')

plt.subplot(339)
plt.plot(df['Size_in_inches'])
plt.title('Size')
```
